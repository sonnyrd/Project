#### I. Import Library

```python
# Import Library
import pandas as pd
import numpy as np
from scipy import stats

# split data into train and test
from sklearn.model_selection import train_test_split

# feature scaling & encoding
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# imputer
from sklearn.impute import SimpleImputer

# pipeline
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# preprocessing
from sklearn.preprocessing import RobustScaler,PolynomialFeatures, StandardScaler

# Modeling
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import GradientBoostingRegressor ,RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

from sklearn.model_selection import cross_val_score, GridSearchCV

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

### III. Data Preprocessing

```python
df.dtypes
```

Kita dapat mengetahui kolom 'WS_TON' dan 'CH_LITER' memiliki dtypes object, dimana jika dilihat pada dataframe column tersebut seharusnya memiliki dtypes numerik (int/float). maka kolom tersebut harus di check

##### Invalid Values

```python
df[' CH_LITER '].replace(' - ',np.nan,inplace=True)
df[' WS_TON '].replace(' - ',np.nan,inplace=True)
```

```python
df[' CH_LITER '] = df[' CH_LITER '].str.replace(r'\D', '')
df[' WS_TON '] = df[' WS_TON '].str.replace(r'\D', '')
```

```python
df[' CH_LITER '] = pd.to_numeric(df[' CH_LITER '],errors='coerce')
df[' WS_TON '] = pd.to_numeric(df[' WS_TON '],errors='coerce')
```

##### Drop Missing Values

```python
df.dropna(inplace=True)
```

##### Drop Date Column

```python
df.drop('DATE',axis=1,inplace=True)
```

##### Drop Duplicated Value

```python
# mencari data yang duplikat
df.duplicated().sum()
```

```python
# reset index dan menghapus data duplikat
df = df.drop_duplicates(keep='first')
df.reset_index(drop = True, inplace = True)
```

##### Model Inference

```python
# get inference set
random = 12
df_inf = df.sample(50,random_state=random)

# drop inference set from dataset
df_train_test = df.drop(df_inf.index)


# reset inference set index
df_inf.reset_index(drop=True, inplace=True)
df_inf.head(10)
```

##### Splitting Dataset

```python
# split data to X and y
X = df_train_test.drop(['CV(AR)'], axis=1)
y = df_train_test['CV(AR)']
```

```python
# split X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=random)

# X_train shape
print('X_train shape: ', X_train.shape)

# X_test shape
print('X_test shape: ', X_test.shape)
```

##### Feature Engineering (Scaling & Encoding)

```python
cat = [col for col in X_train.columns if X_train[col].dtypes == 'O']
num = [col for col in X_train.columns if X_train[col].dtypes != 'O']
print('Numeric Columns:', num)
print('categorical Columns:', cat)
```

```python
numerical_pipeline = Pipeline([
('imputer', SimpleImputer(strategy='constant',fill_value=0)),
('scaler', RobustScaler())
])

categorical_ohe_pipeline = Pipeline([
('imputer', SimpleImputer(strategy='most_frequent')),
('onehot', OneHotEncoder())
])

```

```python
pipeline_fe =  ColumnTransformer([
    ('numeric', numerical_pipeline,num),
    ('category_ohe', categorical_ohe_pipeline,cat),
])
```

### IV. Model Definition

##### I. Linear Regression

```python
# baseline model
linreg = LinearRegression()

#pipeline model
pipe_linreg = Pipeline([
    ('pipeline_fe',pipeline_fe),
    # ('robust',RobustScaler()),
    ('linreg',linreg)
])

#Hyperparameter tuning
param_linreg = {
}
```

##### II. ElasticNet

```python
# baseline model
enet = ElasticNet(random_state=random)

# pipeline model
pipe_enet = Pipeline([
    ('pipeline_fe',pipeline_fe),
    # ('robust',RobustScaler()),
    ('enet',enet)
])

# Hyperparameter tuning
param_enet = {
    'enet__alpha':[0.01,0.1,0.5],
    'enet__l1_ratio':[0.01,0.1,0.5,0.9,1]
}
```

##### III. Gradient Boost Regressor

```python
# Baseline Model
gboost = GradientBoostingRegressor(loss='huber',criterion='squared_error',random_state=random)


# pipeline model
pipe_gboost = Pipeline([
    ('pipeline_fe',pipeline_fe),
    ('RobustScaler',RobustScaler()),
    ('gboost',gboost)
])

# Hyperparameter tuning
param_gboost = {
    'gboost__n_estimators':[200,250,350,400], #default 100
    'gboost__learning_rate':[0.75,0.1,1.25], #default 0.1
    'gboost__max_depth':[2], #default 3
    'gboost__min_samples_split':[5,6,7,8], #default 2
    'gboost__min_samples_leaf':[1,2,3] # default 1


}
```

##### IV. XGBoost Regressor

```python
# Baseline Model
xgb = XGBRegressor()

# pipeline Model
pipe_xgb = Pipeline([
    ('pipeline_fe',pipeline_fe),
    # ('robust',RobustScaler()),
    ('xgb',xgb)
])

# Hyperparameter tuning
param_xgb = {
    'xgb__learning_rate':[0.01,0.05,0.1,1,1.5],
    'xgb__max_depth':[1],
    'xgb__n_estimators':[300,400,500,600]
}
```

##### V. Random Forest Regressor

```python
# Baseline Model
rfr = RandomForestRegressor(random_state=random,n_jobs=-1)

# Pipeline Model
pipe_rfr = Pipeline([
    ('pipeline_fe',pipeline_fe),
    # ('robust',RobustScaler()),
    ('rfr',rfr)   
])

# Hyperparameter Tuning
param_rfr = {
    'rfr__n_estimators':[500,600],
    'rfr__max_depth':[6,7],
    'rfr__min_samples_split':[50]

}
```

### V. Model Training

##### Cross Validation each model

```python
# cross validation model

model_list = [pipe_linreg,pipe_enet,pipe_gboost,pipe_xgb,pipe_rfr]

model_name_list = ['Linreg','ElasticNet','GBoost','XGB','RFR']
mean = []
std   = []
test_score = []

# for loop to get the accuracy of each model
for i in model_list:
    i.fit(X_train,y_train)
    cv = -cross_val_score(i,X_train,y_train,cv=3,scoring='neg_mean_absolute_error',verbose=1)
    mean.append(cv.mean())
    std.append(cv.std())
    i_predict = i.predict(X_test)
    test_score.append(mean_absolute_error(y_test,i_predict))


# create dataframe
cv_result = pd.DataFrame({'model':model_name_list,'Train_Mean':mean,'std':std,'Test_Score':test_score})
```
